```python
import pandas as pd
import numpy as np
import requests
import time

# https://www.kaggle.com/datasets/fronkongames/steam-games-dataset/data
```

```python
columns = ['AppID', 'Name', 'Release date', 'Estimated owners', 'Peak CCU',
    'Required age', 'Price', 'DLC count', 'DiscountDLC count',
    'About the game', 'Supported languages', 'Full audio languages',
    'Reviews', 'Header image', 'Website', 'Support url', 'Support email',
    'Windows', 'Mac', 'Linux', 'Metacritic score', 'Metacritic url',
    'User score', 'Positive', 'Negative', 'Score rank', 'Achievements',
    'Recommendations', 'Notes', 'Average playtime forever',
    'Average playtime two weeks', 'Median playtime forever',
    'Median playtime two weeks', 'Developers', 'Publishers',
    'Categories', 'Genres', 'Tags', 'Screenshots', 'Movies']

Steam_Data = pd.read_csv('/Users/vickywu/Documents/ECO 590 Data Analytics/assignment-1-SakurairoKaden/Steam Games Dataset/games.csv',
    names=columns,
    header=None,
    low_memory=False)

Steam_Data = Steam_Data.iloc[1:].reset_index(drop=True)
```

```python
Steam_Data
```

```python
# Converts the columns, making sure any I didn't crash my laptop
Steam_Data['Positive'] = pd.to_numeric(Steam_Data['Positive'], errors='coerce')
Steam_Data['Negative'] = pd.to_numeric(Steam_Data['Negative'], errors='coerce')
Steam_Data['Price'] = pd.to_numeric(Steam_Data['Price'], errors='coerce')
Steam_Data['Release date'] = pd.to_datetime(Steam_Data['Release date'], errors='coerce')

Steam_Data = Steam_Data.dropna(subset=['Positive', 'Negative', 'Release date'])   # Remove any missing number 
Steam_Data = Steam_Data[(Steam_Data['Positive'] + Steam_Data['Negative']) > 0]    # Removing any games that had no review 
Steam_Data = Steam_Data.reset_index(drop=True)
```

```python
print(f"Rows remaining: {Steam_Data.shape[0]}")
print(f"Release date dtype: {Steam_Data['Release date'].dtype}")
```

```python
## pd.set_option('display.max_rows', None)
## Steam_Data[['Name', 'Release date', 'Positive', 'Negative', 'Price']]
```

```python
Steam_Data['positive_review_ratio'] = (Steam_Data['Positive'] / (Steam_Data['Positive'] + Steam_Data['Negative'])) # Getting a ratio by combining all review together
Steam_Data['release_year'] = Steam_Data['Release date'].dt.year      # To Extract out just the year from the release date

print(f"release_year range: {Steam_Data['release_year'].min()} to {Steam_Data['release_year'].max()}")
Steam_Data[['Name', 'Release date', 'release_year', 'positive_review_ratio']].head(201)
```

```python
# API Scrape - All Time Peak Players (Steam API key) Uses ISteamUserStats via Steam Store API

API_KEY = "0ED2CC3BE4F097A3C98BAC5BCD77BA34"     #Steam Key

def get_peak_players(appid):
    url = "https://store.steampowered.com/appreviews/"
    try:
        r = requests.get(f"https://api.steampowered.com/ISteamUserStats/GetNumberOfCurrentPlayers/v1/",
            params={'key': API_KEY, 'appid': appid},
            timeout=10)
        data = r.json()
        if data['response']['result'] == 1:
            return data['response']['player_count']
    except:
        pass
    return None

sample = Steam_Data.head(500).copy() # Scraping first 500 games for more data

print("Starting API scrape... this will take about 5 minutes")

sample['peak_players'] = sample['AppID'].apply(lambda x: (time.sleep(0.5) or get_peak_players(x)))

print(f"Done! Scraped {sample['peak_players'].notna().sum()} games successfully")
sample[['Name', 'release_year', 'positive_review_ratio', 'peak_players']].sort_values('peak_players', ascending=False).head(101)
```

```python
# Step 6: To Merge API Data into the Datasets
Steam_Data_Final = Steam_Data.merge(sample[['AppID', 'peak_players']], 
    on='AppID', 
    how='left')

Steam_Data_Final['Price'] = pd.to_numeric(Steam_Data_Final['Price'], errors='coerce')
Steam_Data_Final = Steam_Data_Final.reset_index(drop=True)

# Checking to make sure the everything is look correct
print(f"Price range: ${Steam_Data_Final['Price'].min()} to ${Steam_Data_Final['Price'].max()}")
print(f"Missing Price values: {Steam_Data_Final['Price'].isna().sum()}")
print(f"Final dataset shape: {Steam_Data_Final.shape[0]} rows, {Steam_Data_Final.shape[1]} columns")

Steam_Data_Final[['Name', 'release_year', 'positive_review_ratio', 'peak_players', 'Price']].head(201)
```

```python
# Step 7: Save Final Clean Dataset

Steam_Data_Final.to_csv('/Users/vickywu/Documents/ECO 590 Data Analytics/assignment-1-SakurairoKaden/steam_games_final.csv',index=False)

print("File saved successfully!")
print(f"Final dataset: {Steam_Data_Final.shape[0]} rows, {Steam_Data_Final.shape[1]} columns")
print(f"\nVariables in final dataset:")
print(Steam_Data_Final.columns.tolist())
```
